```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns 
import warnings
warnings.filterwarnings('ignore')
```

```python
# 1.读取数据
data = pd.read_excel('./data.xlsx')

# 2.提取所需数据
data = data[data['iUSER_ID']==33]  
data = data[data['iFLOW_NODE_NO']=='自检全检']

# 3.绘制桑基图观察模式
fig= plt.subplots(figsize=(10,8))
plt.subplot(211)

plt.plot(data['dUPDATE_TIME'],data['dUPDATE_TIME'],'ro')  

plt.subplot(212)

plt.plot(data['dNODE_TIME'],data['dNODE_TIME'],'bo')

plt.savefig('模式分析图.png')
```

```python
# 4.总结可能模式
modes = ['集中领取集中提交','分批领取分批提交','囤货模式']

# 5.提取实例
mode1_data = data[data['sBatch_number']==500]
mode2_data = data[data['sBatch_number']==600]  
mode3_data = data[data['iUSER_ID']==45]

# 6.绘示意图
fig,(ax1,ax2) = plt.subplots(figsize=(10,4))
ax.plot(mode1_data)#模式1
ax.plot(mode2_data)#模式2
plt.savefig('模式示意图.png')

# 7.保存结果
results = pd.DataFrame(columns=modes)
results.loc[0] = [mode1_data,"",""]
results.loc[1] = ["",mode2_data,""] 
results.to_excel('结果3.xlsx',index=False)
```
