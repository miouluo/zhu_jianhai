```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from pylab import mpl
mpl.rcParams["font.sans-serif"]=["SimHei"]
mpl.rcParams["axes.unicode_minus"]=False
```

```python
num = np.arange(1,101)
#创建多子图
#subplots(行数、列数)会返回2个数据，fig 表示的是一个画布对象，axes表示的是子图集合
fig,axes=plt.subplots(2,2)
#表示的是在第0行第0列的区域绘图
axes[0,0].plot(num,-num)
axes[0,1].plot(num,-num)
axes[1,0].plot(num,num**2)
axes[1,1].plot(num,np.log(num))
plt.show()
```

```python
#创建单子图
#plt.subplot(行，列，区域索引)区域索引从1开始，不是从0开始
#将画布分为2行2列，并在编号为一的区域绘图
plt.subplot(2,2,1)
plt.plot(num,num)
#将画布分为2行2列，并在编号为二的区域绘图
plt.subplot(2,2,2)
plt.plot(num,-num)
#将画布分为2行1列，并在编号为二的区域绘图
plt.subplot(2,1,2)
plt.plot(num,num**2)
plt.show()
```

# 1.柱状图位置

```python
x_data = ['北京市','上海市','天津市','重庆市']
y_data = np.random.randint(10000,14000,size=4,)
plt.bar(x_data,y_data,color='pink')
plt.title('直辖市GDP')
plt.xlabel('直辖市')
plt.ylabel('GDP值')
#添加柱子上的数字，x，y,字符串
for i,j in zip(x_data,y_data):
    plt.text(i,j+100,j,
             #让字符居中
             ha='center')  
plt.show()
```

```python
#绘制水平条形图
x_data1 = ['亚马逊','当当网','中国图书网','京东','天猫']
y_data1 = [38.8,40.2,48.4,39.9,32.24]
#修改画布形状
plt.figure(figsize=(10,5))
plt.barh(x_data1,y_data1)
plt.title('不同平台书价对比')
plt.xlabel('书价')
plt.ylabel('平台名称')
#添加柱子的数值
for i,j in zip(y_data1,x_data1):
    #第一个参数：需要添加的字符串的x坐标值
    #第二个参数：需要添加的字符串的y坐标值
    #第三个参数：需要添加的字符串
    plt.text(i+0.5,j,i)
    
plt.show()
```

```python
#饼图
labels = ['中专','大专','本科','硕士','其他学历']
x = [0.2515,0.3724,0.3363,0.0368,0.0057]
plt.pie(x,
        #设置标签
        labels=labels,
        explode=[0,0.05,0,0,0],
        #设置百分比
        autopct='%.2f%%',
      # colors=['red','green','gray','yellow','pink']
       )
plt.show()
```
